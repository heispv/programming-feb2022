> Mitoonim be komake argument e `explode` biaim va yek seri az in tike haye data ro az pie chartemoon joda konim. Baraye in kar bayad yek list e jadid tarif konim va be tedade dade hayi ke darim dakhelesh adad bezarim. har adad namayan gare ine ke oon tike az data cheghadr az markaze dayerei ke darim fasele dare.

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.1, 0, 0.3, 0]

plt.pie(y, labels = mylabels, explode = myexplode, startangle = 90)
plt.show() 
```

> Mitoonim be komake argument e `shadow` be pie chartemoon saye ezafe konim.

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.1, 0, 0.3, 0]

plt.pie(y, labels = mylabels, explode = myexplode, startangle = 90, shadow = True)
plt.show() 
```

> baraye inke biaim va range har wedge ro be soorate khas baraye khodemoon moshakhas konim, mitoonim biaim va yek liste jadid besazim va dakhelesh be tedade wedge hayi ke darim rang moshakhas konim. Mitoonim biaim va as esme rang ha, ha keyword rang ha, ya az hexadecimal color oona estefade konim ta range har wedge moshakhas beshe. va oonaro be argument e `color` pass midim.

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.1, 0, 0.3, 0]
wedgecolors = ['#0079FF', '#00DFA2', '#F6FA70', '#FF0060']

plt.pie(y, labels = mylabels, explode = myexplode, startangle = 90, shadow = True, colors = wedgecolors)
plt.show() 
```

> mitoonim az `plt.legend()` estefade konim ta moshakhas konim ke har rang az nemoodari ke darim mokhtas be che data i hast, bayad deghat konim ke agar mikhaim baraye plt.plot az in ravesh estefade konim bayad label haye madde nazaremoon behesh pass bedim.

```python
y1points = np.array([1, 5, 2, 10])
x1 = np.array([1, 2, 6, 7])
y2points = np.array([13, 7, 1, 6])
x2 = np.array([2, 3, 4, 5])

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'green','size':10}


plt.plot(x1, y1points, 'o', ms=5, mec='cornflowerBlue', mfc='cornflowerBlue', ls='-.', color='#6BE51B', lw=2)
plt.plot(x2, y2points, '*', ms=7, mec='#DD581A', mfc='#DD581A', ls='--', color='#6D0000', lw=1)

plt.title('Our Beautiful Plot', fontdict=font2)
plt.xlabel('X', fontdict=font1)
plt.ylabel('Y', fontdict=font1)

plt.legend(['1st', '2nd'])
plt.show()
```

> Be in nokte tavajoh konim ke agar oomadim va be komake `plt.pie` mylabels ro be argument `labels` pass dadim, mitoonim az `plt.legend()` estefade konim chonke amalan aln plt midoone ke label hayi ke darim baraye kodoom data estefade shode, ama agar `labels` ro tooye `plt.pie` pass nadadim dge bayad hatman az `plt.legend(labels)` estefade konim ta bedoone ke legend haye ma chi hastesh.

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.1, 0, 0.3, 0]
wedgecolors = ['#0079FF', '#00DFA2', '#F6FA70', '#FF0060']

plt.pie(y, explode = myexplode, startangle = 90, shadow = True, colors = wedgecolors)
plt.legend(mylabels, loc='upper right', title='Different Fruits')
plt.show() 
```
