# Gain a deeper understanding of GitHub

When it comes to connecting to `GitHub`, there are two main ways to do it:


using the `HTTPS` or `SSH` protocol. Let's take a closer look at each of these options.

1. The `HTTPS` protocol is the simplest way to connect to GitHub. It is a secure protocol that requires you to sign in to your GitHub account before you can read or write any files. This authentication process ensures that your activity on GitHub is recorded in your account history, and that your account remains secure. However, the downside of using `HTTPS` is that you need to enter your GitHub username and password every time you want to access your account, which can become tedious and less secure if your password is weak or reused on multiple platforms.

2. The `SSH` protocol is a more secure way to connect to GitHub. Instead of entering your GitHub username and password every time you want to access your account, you create a public ssh-rsa key on your computer and paste it into your GitHub settings in the `SSH` part. This way, you can authenticate yourself without typing in your credentials each time. The `SSH` protocol also encrypts all data transmission, making it more secure than `HTTPS`. Furthermore, using `SSH` can be especially useful for developers who frequently interact with repositories on GitHub, as it allows for a quicker and smoother workflow.

It's worth noting that while `SSH` is more secure than `HTTPS`, it can be more difficult to set up. It requires some knowledge of the command line and generating `SSH` keys. Therefore, if you're new to GitHub, using `HTTPS` may be the simpler option to start with.

The choice between using the `HTTPS` or `SSH` protocol to connect to GitHub depends on your individual needs and level of experience. `HTTPS` is easy to set up and is secure enough for most users, while `SSH` provides a more secure and efficient workflow for developers who work on GitHub frequently.
