```python
from bs4 import BeautifulSoup
import pandas as pd
import http.cookiejar
from urllib.request import build_opener, HTTPCookieProcessor, Request, urlopen

##############################################################################################################

# Function to get the HTML content from a specified URL and return the BeautifulSoup object
# For test

def get_gainers_loser_soup():
    """
    Fetches the HTML content from a specified URL and returns the BeautifulSoup object.

    Returns:
    - soup (BeautifulSoup): The BeautifulSoup object representing the parsed HTML content.
    """
    # Set a detailed user-agent string
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'}

    # Create a cookie jar and build an opener with cookie handling
    cookie_jar = http.cookiejar.CookieJar()
    opener = build_opener(HTTPCookieProcessor(cookie_jar))

    # Create a request with the specified headers
    req = Request(url='https://www.coingecko.com/en/crypto-gainers-losers', headers=headers)

    try:
        # Open the URL using the opener
        webpage = opener.open(req).read()

    except Exception as e:
        print(f"Error: {e}")

    with urlopen(req) as response:
        html_content = response.read()

    soup = BeautifulSoup(html_content, 'html.parser')
    return soup
```
