```python
y1points = np.array([1, 5, 6, 10])
x1 = np.array([1, 2, 3, 4])
y2points = np.array([13, 7, 5, 6])
x2 = np.array([2, 3, 4, 5])


plt.plot(x1, y1points, 'o', ms=5, mec='cornflowerBlue', mfc='cornflowerBlue', ls='-.', color='#6BE51B', lw=2)
plt.plot(x2, y2points, '*', ms=7, mec='#DD581A', mfc='#DD581A', ls='--', color='#6D0000', lw=1)

plt.show()
```

```python
y1points = np.array([1, 5, 6, 10])
x1 = np.array([1, 2, 3, 4])
y2points = np.array([13, 7, 5, 6])
x2 = np.array([2, 3, 4, 5])


plt.plot(x1, y1points, 'o', ms=5, mec='cornflowerBlue', mfc='cornflowerBlue', ls='-.', color='#6BE51B', lw=2)
plt.plot(x2, y2points, '*', ms=7, mec='#DD581A', mfc='#DD581A', ls='--', color='#6D0000', lw=1)

plt.xlabel('X')
plt.ylabel('Y')

plt.show()
```

Ma mitoonim be komake `plt.title` biaim va title be nemoodaremoon ezafe konim.

```python
y1points = np.array([1, 5, 2, 10])
x1 = np.array([1, 2, 6, 7])
y2points = np.array([13, 7, 1, 6])
x2 = np.array([2, 3, 4, 5])

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'green','size':10}


plt.plot(x1, y1points, 'o', ms=5, mec='cornflowerBlue', mfc='cornflowerBlue', ls='-.', color='#6BE51B', lw=2)
plt.plot(x2, y2points, '*', ms=7, mec='#DD581A', mfc='#DD581A', ls='--', color='#6D0000', lw=1)

plt.title('Our Beautiful Plot', fontdict=font2)
plt.xlabel('X', fontdict=font1)
plt.ylabel('Y', fontdict=font1)

plt.show()
```

```python
# Define the constants
y1points = np.array([1, 5, 2, 10])
x1 = np.array([1, 2, 6, 7])
y2points = np.array([13, 7, 1, 6])
x2 = np.array([2, 3, 4, 5])

# Define the figure size
plt.figure(figsize=(12, 5))

# Define the fonts
font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'green','size':10}

# First Plot
plt.subplot(1, 2, 1)
plt.plot(x1, y1points, 'o', ms=5, mec='cornflowerBlue', mfc='cornflowerBlue', ls='-.', color='#6BE51B', lw=2)
plt.title('Our First Beautiful Plot', fontdict=font1, loc='center')
plt.xlabel('X old', fontdict=font2)
plt.ylabel('Y old', fontdict=font2)

# Second Plot
plt.subplot(1, 2, 2)
plt.plot(x2, y2points, '*', ms=7, mec='#DD581A', mfc='#DD581A', ls='--', color='#6D0000', lw=1)
plt.title('The Second Beautiful Plot', fontdict=font2, loc='right')
plt.xlabel('X', fontdict=font1)
plt.ylabel('Y', fontdict=font1)
plt.grid()

plt.show()
```

Mitoonim be komake `suptitle` yek title kolliam baraye hardo nemoodar benvisim.

```python
# Define the constants
y1points = np.array([1, 5, 2, 10])
x1 = np.array([1, 2, 6, 7])
y2points = np.array([13, 7, 1, 6])
x2 = np.array([2, 3, 4, 5])

# Define the figure size
plt.figure(figsize=(12, 5))

# Define the fonts
font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'green','size':10}
font3 = {'family':'serif','color':'black','size':50}

# First Plot
plt.subplot(1, 2, 1)
plt.plot(x1, y1points, 'o', ms=5, mec='cornflowerBlue', mfc='cornflowerBlue', ls='-.', color='#6BE51B', lw=2)
plt.title('Our First Beautiful Plot', fontdict=font1, loc='center')
plt.xlabel('X old', fontdict=font2)
plt.ylabel('Y old', fontdict=font2)

# Second Plot
plt.subplot(1, 2, 2)
plt.plot(x2, y2points, '*', ms=7, mec='#DD581A', mfc='#DD581A', ls='--', color='#6D0000', lw=1)
plt.title('The Second Beautiful Plot', fontdict=font2, loc='right')
plt.xlabel('X', fontdict=font1)
plt.ylabel('Y', fontdict=font1)
plt.grid()

plt.suptitle('Our Plots', fontdict=font1)
plt.show()
```
