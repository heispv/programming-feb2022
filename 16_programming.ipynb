```python
def test_function(input_1, input_2, input_3=10):
  print(f'input_1: {input_1}, input_2: {input_2}')
  print(input_3)
```

```python
test_function(input_3='you changed it', input_2='input_111', input_1='input_222')
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

Screenshot (12).png

```python
xpoints = np.array([1.5, 14])
ypoints = np.array([7, 15])

plt.plot(xpoints, ypoints)
plt.show()
```

```python
xpoints = np.array([1.5, 14])
ypoints = np.array([7, 15])

plt.plot(xpoints, ypoints, 'o')
plt.show()
```

```python
xpoints = np.array([1.5, 5, 8, 9, 15])
ypoints = np.array([7, 15, 10, -3, 2])

plt.plot(xpoints, ypoints, marker='*')
plt.show()
```

```python
xpoints = np.array([1.5, 5, 8, 9, 15])
ypoints = np.array([7, 15, 10, -3, 2])

plt.plot(xpoints, ypoints, marker='|')
plt.show()
```

```python
xpoints = np.array([1.5, 5, 8, 9, 15])
ypoints = np.array([7, 15, 10, -3, 2])

plt.plot(xpoints, ypoints, marker='v')
plt.show()
```

```python
xpoints = np.array([1.5, 5, 8, 9, 15])
ypoints = np.array([7, 15, 10, -3, 2])

plt.plot(xpoints, ypoints, 'v-.b')
plt.show()
```

```python
xpoints = np.array([1.5, 5, 8, 9, 15])
ypoints = np.array([7, 15, 10, -3, 2])

plt.plot(xpoints, ypoints, 'p:y')
plt.show()
```

```python
xpoints = np.array([1.5, 5, 8, 9, 15])
ypoints = np.array([7, 15, 10, -3, 2])

plt.plot(xpoints, ypoints, '.:c', ms=20)
plt.show()
```
