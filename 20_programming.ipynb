write me python code. I want you to generate 100 random x and y numbers and store them in a separate list. then use matplotlib library and plot a scatter plot using this random generated numbers. I want them to have a heatmap and their values should be random too.
`

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate random x and y coordinates
np.random.seed(42)  # Set seed for reproducibility
num_points = 100
x = np.random.rand(num_points)
y = np.random.rand(num_points)

# Generate random values for the heatmap
heatmap_values = np.random.rand(num_points)

# Plot scatter plot with heatmap
plt.scatter(x, y, c=heatmap_values, cmap='hot', alpha=0.7)
plt.colorbar()  # Add colorbar for heatmap
plt.title('Scatter Plot with Heatmap')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate random x and y coordinates
np.random.seed(42)  # Set seed for reproducibility
num_points = 100
x = np.random.rand(num_points)
y = np.random.rand(num_points)

# Generate random values for the heatmap
heatmap_values = np.random.rand(num_points)

# Plot scatter plot with heatmap
plt.scatter(x, y, c=heatmap_values, cmap='hot', alpha=0.7)
colorbar = plt.colorbar(orientation='horizontal')  # Add horizontal colorbar
colorbar.set_label('Heatmap Values')  # Set label for the colorbar
plt.title('Scatter Plot with Heatmap')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate random x and y coordinates
np.random.seed(42)  # Set seed for reproducibility
num_points = 100
x = np.random.rand(num_points)
y = np.random.rand(num_points)

# Generate random values for the heatmap
heatmap_values = np.random.rand(num_points)

# Plot scatter plot with heatmap
fig, ax = plt.subplots()
scatter = ax.scatter(x, y, c=heatmap_values, cmap='hot', alpha=0.7)

# Add horizontal colorbar
colorbar = plt.colorbar(scatter, orientation='horizontal')
colorbar.set_label('Heatmap Values')

# Set title with blue font color and font size of 15
ax.set_title('Scatter Plot with Heatmap', color='blue', fontsize=15)

ax.set_xlabel('X')
ax.set_ylabel('Y')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate random x and y coordinates
np.random.seed(42)  # Set seed for reproducibility
num_points = 100
x = np.random.rand(num_points)
y = np.random.rand(num_points)

# Generate random values for the heatmap
heatmap_values = np.random.rand(num_points)

# Plot scatter plot with heatmap
plt.scatter(x, y, c=heatmap_values, cmap='hot', alpha=0.7)
plt.colorbar(label='Heatmap Values', orientation='horizontal')  # Add colorbar with label

plt.title('Scatter Plot with Heatmap', color='blue', fontsize=15)  # Set title with blue font color and size 15
plt.xlabel('X')
plt.ylabel('Y')

plt.show()
```
